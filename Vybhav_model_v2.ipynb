```python
import numpy as np 
import pandas as pd 

%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading the dataset
df = pd.read_csv("data.csv")
df.head()
```

#### The features represent the mean, std, and worst values for radius, texture, perimeter, area, etc. In this analysis, we will consider only the mean values of the features.

```python
#Dropping the ID and NaN values columns
df.drop('id',axis=1,inplace=True)
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
# Converting the output feature into a numerical mapped one.
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# Split the data into features and target variable
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Normalize the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Create the model
log_reg_model = LogisticRegression(random_state=42, max_iter=1000)
```

```python
# Train the model
log_reg_model.fit(X_train_scaled, y_train)
```

```python
# Predict the target for the test set
y_pred = log_reg_model.predict(X_test_scaled)
```

```python
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
# Plotting the confusion matrix
plt.figure(figsize=(7, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
